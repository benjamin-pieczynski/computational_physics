```python
# import statements
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# load in data function
def load_data(file_name, skip, pack):
    # loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
    # we use skiprows=1 to avoid the header
    all_data = np.loadtxt(file_name,skiprows=skip, unpack=pack)
    
    return all_data
```

```python
# basic analytic width
sigma = []
sigma_0 = 0.5
h = 1
m = 1
time = np.linspace(0,5, 100)
for t in time:
    sigma.append(np.sqrt(sigma_0**2 + (h**4)*(t**2)/((4*(m**2)*sigma_0**2))))
```

```python
# time results
t_r = load_data('time_results.dat', 1, True)

t = t_r[0]
norm = t_r[1]
pos = t_r[2]
width = t_r[3]
    
#plt.plot(t, width)
plt.plot(t, width)
plt.plot(time, sigma)
plt.xlabel('Time(s)')
plt.ylabel('\u03C3')
plt.title('Width over time - Basic')
plt.gca().legend(('numerical result','analytic result'), bbox_to_anchor = (1,1))
plt.show()
```

It isn't exactly like the analytic result, but it is close

```python
# advanced x position

k = 1 # used k_oscillator = 1
x_0 = 0.2 # used center around 0.2
x_t = []
t_x = np.linspace(0,5,100)
for ti in t_x:
    x_t.append(x_0 * np.cos(np.sqrt(k/m)*ti))
```
