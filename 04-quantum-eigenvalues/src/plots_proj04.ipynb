```python
# import statements
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# load in data function
def load_data(file_name):
    # loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
    # we use skiprows=1 to avoid the header
    all_data = np.loadtxt(file_name,skiprows=1)
    # We transpose the array in order to be able to extract each column individually
    all_data = all_data.transpose()
    
    return all_data
    
```

```python
# loading in all the data

# inf well
file_inf_well = 'infinite_well_wf.dat'
data_inf_well = load_data(file_inf_well)
x_inf = data_inf_well[0]
inf_ground = data_inf_well[1]
inf_1st = data_inf_well[2]
inf_2nd = data_inf_well[3]

# harmonic oscillator
file_harm_os = 'harmonic_oscillator_wf.dat'
data_harm_os = load_data(file_harm_os)
x_harm = data_harm_os[0]
harm_ground = data_harm_os[1]
harm_1st = data_harm_os[2]
harm_2nd = data_harm_os[3]

# woods saxon
file_ws = 'woods_saxon_wf.dat'
data_ws = load_data(file_ws)
x_ws = data_ws[0]
ws_ground = data_ws[1]
ws_1st = data_ws[2]
ws_2nd = data_ws[3]

# woods saxon ener
file_ws2 = 'woods_saxon_ener.dat'
data_ws2 = load_data(file_ws2)
r_ws2 = data_ws2[0]
ws_ground2 = data_ws2[1]
ws_1st2 = data_ws2[2]
ws_2nd2 = data_ws2[3]
```

**Plot 1: Ground State Normalized Probability Density**

The ground state normalized probability density for the 3 different problems

```python
plt.plot(x_inf, inf_ground)
plt.plot(x_harm, harm_ground)
plt.plot(x_ws, ws_ground)
plt.xlabel('X (fm)')
plt.ylabel('Normalized Probability Density')
plt.title('Normalized Probability Density: Ground State')
plt.gca().legend(('Infinite Well','Harmonic Oscillator', 'Woods-Saxon'), bbox_to_anchor = (1,1))
plt.show()
```

**Plot 2: 1st Excited State Normalized Probability Density**

The 1st excited state normalized probability density for the 3 different problems

```python
plt.plot(x_inf, inf_1st)
plt.plot(x_harm, harm_1st)
plt.plot(x_ws, ws_1st)
plt.xlabel('X (fm)')
plt.ylabel('Normalized Probability Density')
plt.title('Normalized Probability Density: 1st Excited State')
plt.gca().legend(('Infinite Well','Harmonic Oscillator', 'Woods-Saxon'), bbox_to_anchor = (1,1))
plt.show()
```

**Plot 3: The 2nd excited state normalized probability density**

The 2nd excited state normalized probability density for the 3 different problems

```python
plt.plot(x_inf, inf_2nd)
plt.plot(x_harm, harm_2nd)
plt.plot(x_ws, ws_2nd)
plt.xlabel('X (fm)')
plt.ylabel('Normalized Probability Density')
plt.title('Normalized Probability Density: 2nd Excited State')
plt.gca().legend(('Infinite Well','Harmonic Oscillator', 'Woods-Saxon'), bbox_to_anchor = (1,1))
plt.show()
```

**Plot 4: Woods-Saxon Potential**

The three lowest energies for the Woods-Saxon potential as a function of the radius 

```python
plt.plot(r_ws2, ws_ground2)
plt.plot(r_ws2, ws_1st2)
plt.plot(r_ws2, ws_2nd2)
plt.xlabel('Radius')
plt.ylabel('Normalized Probability Density')
plt.title('Woods Saxon Potential R 2-10')
plt.gca().legend(('ground state','1st Excited State', '2nd Excited State'), bbox_to_anchor = (1,1))
plt.show()
```
